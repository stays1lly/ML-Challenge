```python
import pandas as pd
df = pd.read_csv('train.csv')
print(df.head())
```

```python
import os
import requests
from PIL import Image
import io

# Function to fetch image from a link and open it with PIL
def fetch_image(link):
    try:
        response = requests.get(link, timeout=10)
        response.raise_for_status()  # Check for a valid response
        img = Image.open(io.BytesIO(response.content))
        return img
    except Exception as e:
        print(f"Error fetching image: {e}")
        return None
```

```python
def process_images(df, download_folder):
    # Ensure the folder to save images exists
    if not os.path.exists(download_folder):
        os.makedirs(download_folder)
    
    for idx, image_link in enumerate(df['image_link']):
        # Fetch image
        image = fetch_image(image_link)
        
        if image:
            # Save image locally (optional)
            file_name = f"image_{idx}.jpg"
            file_path = os.path.join(download_folder, file_name)
            image.save(file_path)  # Save image as JPEG
            print(f"Image {idx} saved: {file_name}")
        else:
            print(f"Failed to process image {idx}")
```

```python
process_images(df, 'Images/')
```

```python
import numpy as np
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input

# Resize and preprocess the image
def preprocess_image(image, target_size=(224, 224)):
    # Resize the image to match the input size of the model (e.g., 224x224 for ResNet50)
    image = image.resize(target_size)
    img_array = np.array(image)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = preprocess_input(img_array)  # Preprocess for ResNet50
    return img_array

# Preprocess the image for ResNet50
processed_image = preprocess_image(image)
```

```python
from tensorflow.keras.applications import ResNet50

# Load the ResNet50 model with ImageNet weights, without the top classification layer
resnet_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')

# Extract features from the preprocessed image
features = resnet_model.predict(processed_image)

# The extracted features are a 2048-dimensional vector
print("Extracted Features:", features.shape)
```
